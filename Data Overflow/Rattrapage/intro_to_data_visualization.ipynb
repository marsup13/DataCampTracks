```
import matplotlib.pyplot as plt
import numpy as np
```

```

fig,ax = plt.subplots()
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

```
fig,ax = plt.subplots()
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

```
df2= pd.read_csv("/content/drive/MyDrive//Data_visualization//Coronavirus_disease.csv",index_col=0)
df2.head(5)
```

```
fig,ax=plt.subplots()
ax.bar(df2.index,df2["Deaths"])
plt.show()
```

```
fig,ax=plt.subplots()
ax.bar(df2.index,df2["Deaths"])

ax.set_xticklabels(df2.index,rotation=90)
ax.set_ylabel("counts")
ax.set_xlabel("Countrys")
ax.set_title("covid-19")
plt.show()
```

```
fig,ax=plt.subplots()
ax.bar(df2.index,df2["Deaths"])
ax.bar(df2.index,df2["Recovered"],bottom=df2["Deaths"])
ax.set_xticklabels(df2.index,rotation=90)
ax.set_ylabel("counts")
ax.set_xlabel("Countrys")
ax.set_title("covid-19")
plt.show()
```

```
import seaborn as sns 
```

```
sns.countplot(x=df1["Gender"])
plt.show()
```

```
plot_color= {"Male":"blue", "Female":"red"}
sns.countplot(x=df1["Gender"],palette=plot_color)
plt.show()
```

```
sns.set_palette("GnBu_r")
sns.countplot(x=df1["Gender"])
plt.xticks(rotation=90)
plt.show()
```

```
plot_color= {"Male":"blue", "Female":"red"}
sns.countplot(x=df1["Live_Cairo"],hue=df1["Gender"])
plt.show()
```
